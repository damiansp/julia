```julia
function spherevol(r)
    return 4/3 * π * r^3
end
```

```julia
spherevol(2)
```

```julia
quadratic(a, sqrterm, b) = (-b + sqrterm) / 2a
```

```julia
quadratic(3, 4, 5)
```

```julia
function quadratic2(a::Float64, b::Float64, c::Float64)
    sqrterm = sqrt(b^2 - 4a*c)
    r1 = quadratic(a, sqrterm, b)
    r2 = quadratic(a, -sqrterm, b)
    r1, r2
end
```

```julia
vol = spherevol(3)
@printf "vol: %0.3f" vol
```

```julia
q1, q2 = quadratic2(2., -2., -12.)
print("res: ", q1, " ", q2)
```

### Arrays

```julia
function summarize(a)
    println(summary(a), ": ", repr(a))
end
```

```julia
a1 = [1, 2, 3]
summarize(a1)
```

```julia
a2 = []
summarize(a2)
```

```julia
a3 = Int64[]
summarize(a3)
```

```julia
a4 = 1:8
summarize(a4)
```

```julia
a4 = collect(a4)
summarize(a4)
```

```julia
a5 = [2^i for i = 1:10]
summarize(a5)
```

```julia
a6 = (Array{Int64, 1})[]
summarize(a6)
```

```julia
push!(a1, 4)
summarize(a1)
```

```julia
push!(a3, 1)
summarize(a3)
```

```julia
push!(a6, [1, 2, 3])
summarize(a6)
```

```julia
a7 = repeat(a1, inner=[2], outer=[1])
summarize(a7)
```

```julia
a8 = repeat(collect(4:-1:1), inner=[1], outer=[2])
summarize(a8)
```

```julia
a = collect(1:20)
for i in a
    if i % 2 == 1
        continue
    end
    print(i, ", ")
    if i >= 8
        break
    end
end
```

```julia
function checktype(x)
    print("It's a... ")
    if x isa Int
        println("integer")
    elseif x isa AbstractFloat
        println("float!")
    elseif x isa Complex
        println("imaginary number!")
    else
        println("a... hm.. dunno.")
    end
end
```

```julia
checktype(2)
```

```julia
checktype(√2)
```

```julia
checktype("this guy")
```
