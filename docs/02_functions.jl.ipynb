```julia
function hypot(x, y)
    x = abs(x)
    y = abs(y)
    if x > y
        r = y / x
        return x * sqrt(1 + r*r)
    end
    if y == 0
        return zero(x)
    end
    r = x / y
    return y * sqrt(1 + r*r)
end

hypot(3, 4), hypot(1, 0), hypot(7, 3)
```

```julia
map(x -> begin
        if x < 0 && iseven(x)
            return 0
        elseif x == 0
            return 1
        else
            return x
        end
    end,
    [-1, 0, 1, 2])
```

```julia
# Cleaner
map([-1, 0, 1, 2]) do x
    if x < 0 && iseven(x)
        return 0
    elseif x == 0
        return 1
    else
        return x
    end
end
```
