# Conditional Evaluation

```julia
x = rand()
y = rand()

if x < y
    println("$x < $y")
elseif x > y
    println("$x > $y")
else
    println("$x = $y")
end
```

```julia
x = rand()
y = rand()
println(x < y ? "$x < $y" : "$x ≥ $y")
```

```julia
x = rand()
y = rand()
println(x < y ? "$x < $y" : (x > y ? "$x > $y" : "$x = $y"))
```

# Short-Circuit Evaluation

```julia
function fact(n::Int)
    n >= 0|| error("n cannot be negative")
    n == 0 && return 1
    n * fact(n - 1)
end

fact(5), fact(0)
```

```julia
fact(-3)
```

# Errors

```julia
fussy_sqrt(x) = x >= 0 ? sqrt(x) : error("negative x not allowed")
```

```julia
fussy_sqrt(4)
```

```julia
fussy_sqrt(-4)
```

```julia
function verbose_sqrt(x)
    println("before fussy sqrt")
    r = fussy_sqrt(x)
    println("after")
    return r
end
```

```julia
verbose_sqrt(4)
```

```julia
verbose_sqrt(-4)
```

```julia
f(x) = try
    sqrt(x)
catch
    sqrt(complex(x, 0))
end
```

```julia
f(4)
```

```julia
f(-4)
```

```julia
sqrt_second(x) = try
    sqrt(x[2])
catch e
    if isa(e, DomainError)
        sqrt(complex(x[2], 0))
    elseif isa(e, BoundsError)
        sqrt(x)
    end
end
```

```julia
sqrt_second([1, 4])
```

```julia
sqrt_second([1, -4])
```

```julia
sqrt_second(9)
```
