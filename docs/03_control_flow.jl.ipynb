# Conditional Evaluation

```julia
x = rand()
y = rand()

if x < y
    println("$x < $y")
elseif x > y
    println("$x > $y")
else
    println("$x = $y")
end
```

```julia
x = rand()
y = rand()
println(x < y ? "$x < $y" : "$x ≥ $y")
```

```julia
x = rand()
y = rand()
println(x < y ? "$x < $y" : (x > y ? "$x > $y" : "$x = $y"))
```

# Short-Circuit Evaluation

```julia
function fact(n::Int)
    n >= 0|| error("n cannot be negative")
    n == 0 && return 1
    n * fact(n - 1)
end

fact(5), fact(0)
```

```julia
fact(-3)
```

# Loops

```julia
i = 1

while i <= 5
    println(i)
    global i += 1
end
```

```julia
for j = 1:5
    println(j)
end
```

```julia
for i in [1, 4, 0] 
    println(i)
end
```

```julia
i = 1

while true
    print("$i ")
    if i > 5
        break
    end
    global i += 1
end
```

```julia
for j = 1:1000
    print("$j...")
    if j >= 5
        break
    end
end
```

```julia
for i = 1:10
    if i % 3 != 0
        continue
    end
    print(("//$i\\\\ "))
end
```
