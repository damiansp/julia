# Global Scope

```julia
module A
    a = 1 # a global in A's scope
end


module B
    module C
        c = 2
    end
    
    b = C.c
    import ..A
    d = A.a
end
```

```julia
module D
    b = a # UndefVarError
end
```

```julia
module E
    import ..A
    A.a = 2 # Error: cannot assign
end
```

# Local Scope

```julia
for i = 1:10
    z = i
end

z
```

```julia
for i = 1:3
    z = i
    for j = 4:6
        z = 0
    end
    println(z)
end
```

```julia
for i = 1:3
    x = i + 1
    for j = 4:6
        local x = 0
    end
    println(x)
end
```

```julia
for i = 1:10
    global z
    z = i
end

z
```

```julia
for i = 1:10
    z = i
    global z
end

z
```

```julia
x, y = 1, 2

function foo()
    x = 2
    return x + y # x is local, y is global
end

foo()
```

```julia
x
```

```julia
x = 1

function fubar()
    global x = 2
end

fubar()
```

```julia
x
```

```julia
x, y = 1, 2

function baz()
    x = 2
    function bar()
        x = 10 # modifies parent x (not global)
        return x + y # global y
    end
    return bar() + x # 12 + 10
end

baz()
```

```julia
x, y
```

```julia
let state = 0
    global counter() = (state += 1)
end
```

```julia
function bar()
    x = 10 # local
    return x + y
end
```
