# Defining Methods

```julia
f(x::Float64, y::Float64) = 2x + y
```

```julia
f(2., 3.)
```

```julia
f(2., 3)
```

```julia
f(x::Number, y::Number) = 3x - y
```

```julia
f(2., 4)
```

```julia
f(3, 4.)
```

```julia
f(x, y) = println("Whoa there, Nelly") # args default to Any
```

```julia
f("foo", 1)
```

# Method Ambiguities

```julia
g(x::Float64, y) = 2x + y
```

```julia
g(x, y::Float64) = x + 2y
```

```julia
g(2., 3), g(2, 3.)
```

```julia
g(2., 3.)
```

```julia
g(x::Float64, y::Float64) = 2x + 2y
```

```julia
g(2., 3.)
```

# Parametric Methods

```julia
same_type(x::T, y::T) where {T} = true
```

```julia
same_type(x, y) = false
```

```julia
same_type(1, 2), same_type(1., 2)
```

```julia
same_type("foo", "bar"), same_type(Int32(1), Int64(2))
```

```julia
myappend(v::Vector{T}, x::T) where {T} = [v..., x]
```

```julia
myappend([1, 2, 3], 4)
```

```julia
myappend([1, 2, 3], 4.)
```

```julia
mytypeof(x::T) where {T} = T
```

```julia
mytypeof(1), mytypeof(1.)
```

```julia
same_type_numeric(x::T, y::T) where {T <: Number} = true
```

```julia
same_type_numeric(x::Number, y::Number) = false
```

```julia
same_type_numeric(1, 2), same_type_numeric(1., 2.)
```

```julia
same_type_numeric(1, 2.)
```
