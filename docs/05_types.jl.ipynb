````julia
function sinc(x)::Float64
    if x == 0
        return 1
    end
    return sin(pi * x) / (pi * x)
end
````

# Abstract Types

````julia
# Julia Number hierarchy
abstract type Number end
abstract type Real <: Number end
abstract type AbstractFloat <: Real end
abstract type Integer <: Real end
abstract type Signed <: Integer end
abstract type Unsigned <: Integer end
````

````julia
Integer <: Number
````

````julia
Integer <: AbstractFloat
````

````julia
function myplus(x, y)
    x + y
end
````

````julia
function myplus(x::Int, y::Int)
    x + y
end
````

# Composite Types

````julia
struct Foo
    bar # defaults to type Any
    baz::Int
    qux::Float64
end
````

````julia
foo = Foo("Hello, World!", 23, 1.5)
````

````julia
typeof(foo)
````

````julia
fieldnames(Foo)
````

````julia
foo.bar
````

````julia
foo.baz
````

# Mutable Composite Types

````julia
mutable struct Bar
    baz
    qux::Float64
end
````

````julia
bar = Bar("Hello", 1.5)
bar.qux = 2.
bar.baz = 1 // 2
bar.baz
````

````julia
typeof(Real)
````

# Type Unions

````julia
IntOrString = Union{Int, AbstractString}
1::IntOrString
"Hello"::IntOrString
````

````julia
1.0::IntOrString
````

# Parametric Composite Types

````julia
struct Point{T}
    x::T
    y::T
end
````

````julia
Point{Float64}
````

````julia
Point{AbstractString}
````

````julia
Point{Float64} <: Point
````

````julia
function norm(p::Point{Real})
    sqrt(p.x^2 + p.y^2)
end
````

````julia
Point{Float64}(1., 2.)
````

# Parametric Abstract Types

````julia
abstract type Pointy{T} end
````

````julia
Pointy{Int64} <: Pointy
````

````julia
struct Point{T} <: Pointy{T}
    x::T
    y::T
end
````
