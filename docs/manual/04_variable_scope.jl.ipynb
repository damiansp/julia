# Global Scope

```julia
module A
    a = 1 # a global in A's scope
end


module B
    module C
        c = 2
    end
    
    b = C.c
    import ..A
    d = A.a
end
```

```julia
module D
    b = a # UndefVarError
end
```

```julia
module E
    import ..A
    A.a = 2 # Error: cannot assign
end
```

# Local Scope

```julia
for i = 1:10
    z = i
end

z
```

```julia
for i = 1:3
    z = i
    for j = 4:6
        z = 0
    end
    println(z)
end
```

```julia
for i = 1:3
    x = i + 1
    for j = 4:6
        local x = 0
    end
    println(x)
end
```

```julia
x, y = 1, 2

function foo()
    x = 2
    return x + y # x is local, y is global
end

foo()
```

```julia
x, y = 1, 2

function baz()
    x = 2
    function bar()
        x = 10 # modifies parent x (not global)
        return x + y # global y
    end
    return bar() + x # 12 + 10
end

baz()
```

# Let Blocks

```julia
x, y, z = -1, -1, -1

let x = 1, z
    println("x: $x, y: $y") # local x (1), global y (-1)
    println("z: $z")        # local z exists, but not defined
end
```

```julia
Fs = Vector{Any}(undef, 2)
i = 1

while i <= 2
    Fs[i] = () -> i
    global i += 1
end
```

```julia
Fs[1](), Fs[2]()
```

```julia
Fs = Vector{Any}(undef, 2)
i = 1

while i <= 2
    let i = i
        Fs[i] = () -> i
    end
    global i += 1
end
```

```julia
Fs[1](), Fs[2]()
```

```julia
let 
    local x = 1
    let 
        local x = 2
    end
    x
end
```

# For Loops and Comprehensions

```julia
Fs = Vector{Any}(undef, 2)

for j = 1:2
    Fs[j] = () -> j
end

Fs[1](), Fs[2]()
```

```julia
function f()
    i = 0
    for i = 1:3
    end
    i
end

f()
```

```julia
function f()
    i = 0
    for outer i = 1:3
    end
    i
end

f()
```
