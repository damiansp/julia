```julia
function fib(n)
    (n == 0 || n == 1) && return n
    return fib(n - 1) + fib(n - 2)
end
```

```julia
fib(0), fib(1), fib(7)
```

### 1. Arguments

```julia
add3(a::Int64, b::Int64, c::Int64) = a + b + c
```

```julia
values = 1, 2, 3
```

```julia
function addmean(init, args...)
    s = 0
    for arg in args
        s += arg
    end
    init +  s/length(args)
end
```

```julia
addmean(10, values...)
```

```julia
addmean(10, 10, 20, 37)
```
