### 1. Arguments

```julia
add3(a::Int64, b::Int64, c::Int64) = a + b + c
```

```julia
values = 1, 2, 3
```

```julia
function addmean(init, args...)
    s = 0
    for arg in args
        s += arg
    end
    init +  s/length(args)
end
```

```julia
addmean(10, values...)
```

```julia
addmean(10, 10, 20, 37)
```

### 8. Broadcasting Functions

```julia
f1(a::Int64, b::Int64) = a * b
f1(2, 3)
```

```julia
f1([2, 3], [4, 6])
```

```julia
broadcast(f1, [2, 3], [4, 6])
```

```julia
f1.([2, 3], [4, 6])
```

```julia
f2(a::Int64, b::Int64, c::Array{Int64, 1}, d::Array{Int64, 1}) = (
    a*b + sum(c) - sum(d))
```

```julia
f2(1, 2, [1, 2, 3], [0, 0, 1]) # no broadcasting
```

```julia
# broadcast over 1st 2 args only
f2.([1, 1, 1], [2, 2, 2], Ref([1, 2, 3]), Ref([0, 0, 1]))
```

### 6. Exiting

```julia
function workspace(f, args...)
    println(f(args...))
    exit()
end
```

```julia
workspace(f2, 1, 2, [1, 2, 3], [0, 0, 1])
```
