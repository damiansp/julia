```julia
x = 1:10
y = rand(10, 4)
plot(x, y)
```

```julia
plot(x, y, layout=(2, 2))
```

```julia
p1 = plot(x, y)
p2 = scatter(x, y)
p3 = plot(x, y, xlabel="A Label!", lw=3, title="Subtitle!")
p4 = histogram(x, y)
plot(p1, p2, p3, p4, layout=(2, 2), legend=false)
```

```julia
df = DataFrame(a=1:10, b=10 * rand(10), c=10 * rand(10))
@df df plot(:a, [:b, :c])
```

```julia
plot(df[:, :a], [df[:, :b], df[:, :c]])
```

```julia
@df df scatter(:a, :b, title="DF Scatter")
```

```julia
scatter(df[:, :a], df[:, :b], title="DF Scatter")
```
