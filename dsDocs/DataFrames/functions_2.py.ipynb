```julia
df = DataFrame(a=[1, 1, 1, 2, 2, 1, 1, 2],
               b=repeat([2, 1], outer=4),
               c=1:8)
```

```julia
gd = groupby(df, :a)
select(gd, :c => sum, nrow)
```

```julia
select(gd, :c => sum, nrow, ungroup=false)
```

```julia
select(gd, :c => (x -> sum(log, x)) => :sum_log_c)
```

```julia
select(gd, [:b, :c] .=> sum)
```

```julia
select(gd, :b => :b1, :c => :c1, [:b, :c] => +, keepkeys=false)
```

```julia
select(gd, :b, :c => sum)
```

```julia
select(gd, :, AsTable(Not(:a)) => sum)
```
