```julia
df1 = DataFrame(A=1:3, B=11:13)
df2 = DataFrame(A=4:6, B=14:16)
df3 = hcat(df1, df2, makeunique=true)
```

```julia
df3 = hcat(df1, df2)
```

```julia
df3.A === df1.A
```

```julia
df3 = hcat(df1, df2, makeunique=true, copycols=false)
```

```julia
df3.A === df1.A
```
