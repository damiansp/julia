```julia
df1 = DataFrame(A=1:3, B=11:13)
df2 = DataFrame(A=4:6, B=14:16)
df3 = hcat(df1, df2, makeunique=true)
```

```julia
df3 = hcat(df1, df2)
```

```julia
df3.A === df1.A
```

```julia
df3 = hcat(df1, df2, makeunique=true, copycols=false)
```

```julia
df3.A === df1.A
```

```julia
df = DataFrame(x=[-3, -1, 0, 2, 4], y=1:5)
```

```julia
sort(df, order(:x, rev=true))
```

```julia
sort(df, order(:x, by=abs))
```

```julia
push!(df, (true, false))
```

```julia
push!(df, df[1, :])
```

```julia
push!(df, (C="something", x=true, y=false), cols=:intersect)
```

```julia
push!(df, Dict(:x=>5, :z=>9), cols=:union)
```

```julia
push!(df, NamedTuple(), cols=:subset)
```
