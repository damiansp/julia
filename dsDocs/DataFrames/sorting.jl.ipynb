```julia
using DataFrames
using RDatasets
```

```julia
iris = dataset("datasets", "iris")
first(iris, 3)
```

```julia
sort!(iris)
first(iris, 3)
```

```julia
last(iris, 3)
```

```julia
sort!(iris, rev=true)
first(iris, 3)
```

```julia
sort!(iris, (order(:Species, by=length), order(:SepalLength, rev=true)))
first(iris, 5)
```

```julia
sort!(iris, [:Species, :PetalLength], rev=(true, false))
first(iris, 3)
```

```julia
sort!(iris, [order(:Species, rev=true), :PetalLength])
first(iris, 3)
```
