### Constructing

```julia
df = DataFrame()
df.A = 1:8
df.B = ["M", "F", "M", "M", "F", "M", "F", "F"]
df
```

```julia
size(df), size(df, 1)
```

```julia
df = DataFrame(A=Int[], B=String[])
push!(df, (1, "M"))
push!(df, (2, "N"))
df
```

```julia
push!(df, Dict(:B => "O", :A => 3))
df
```

```julia
df = DataFrame(a=[1, 2, 3], b=[:a, :b, :c])
df
```

```julia
CSV.write("dataframe.csv", df)
```
