```julia
df = DataFrame(x1=[1, 2], x2=[3, 4], y=[5, 6])
transform(df, All() => +)
```

```julia
Random.seed!(3)
df = DataFrame(rand(10, 3), [:a, :b, :c])
```

```julia
transform(df, AsTable(:) => ByRow(argmax) => :prediction)
```

```julia
df = DataFrame(x=[1, 2, missing], y=[1, missing, missing])
transform(
    df, 
    AsTable(:) .=> ByRow.(
        [sum ∘ skipmissing, 
         x -> count(!ismissing, x), 
         mean ∘ skipmissing]
    ) .=> [:sum, :n, :mean])
```
