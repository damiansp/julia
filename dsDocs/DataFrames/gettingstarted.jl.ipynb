### Constructing

```julia
df = DataFrame()
df.A = 1:8
df.B = ["M", "F", "M", "M", "F", "M", "F", "F"]
df
```

```julia
size(df), size(df, 1)
```

```julia
df = DataFrame(A=Int[], B=String[])
push!(df, (1, "M"))
push!(df, (2, "N"))
df
```

```julia
push!(df, Dict(:B => "O", :A => 3))
df
```

### Working with DataFrames

```julia
df = DataFrame(A=1:2:1000, B=repeat(1:10, inner=50), C=1:500)
```

```julia
first(df, 3)
```

```julia
last(df, 3)
```

```julia
DataFrame(
  a=1:2, b=[1.0, missing], c=categorical('a':'b'), de=[1//2, missing])
```

```julia
df[1:3, :]
```

```julia
df[[1, 5, 10], :]
```

```julia
df[:, [:A, :B]]
```

```julia
df[1:3, [:B, :A]]
```

```julia
df[[4, 1], [:C]]
```

```julia
typeof(df[!, [:A]]), typeof(df[:, [:A]])
```

```julia
typeof(df[!, :A]), typeof(df[:, :A])
```

```julia
df = DataFrame(x1=1, x2=2, y=3)
df[!, r"x"]
```

```julia
df[!, Not(:x1)]
```

```julia
df = DataFrame(r=1, x1=2, x2=3, y=4)
df[:, All(r"x", :)]
```

```julia
df[:, All(Not(r"x"), :)]
```

```julia
df = DataFrame(A=1:2:1000, B=repeat(1:10, inner=50), C=1:500)
df[df.A .> 500, :]
```

```julia
df[(df.A .> 500) .& (300 .< df.C .< 400), :]
```

```julia
df[in.(df.A, Ref([1, 5, 601])), :]
```
