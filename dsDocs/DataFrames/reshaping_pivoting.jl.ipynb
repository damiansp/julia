```julia
using CSV, DataFrames, RDatasets, Statistics
```

```julia
iris = dataset("datasets", "iris")
first(iris, 3)
```

```julia
last(iris, 3)
```

```julia
d = stack(iris, 1:4)
first(d, 3)
```

```julia
last(d, 3)
```

```julia
d = stack(iris, [:SepalLength, :SepalWidth, :PetalLength])
last(d, 3)
```

```julia
d = stack(iris, [:SepalLength, :SepalWidth], :Species)
first(d, 3)
```

```julia
d = stack(iris, Not(:Species))
first(d, 3)
```

```julia
iris.id = 1:size(iris, 1)
longdf = stack(iris, Not([:Species, :id]))
first(longdf, 3)
```

```julia
widedf = unstack(longdf, :id, :variable, :value)
first(widedf, 3)
```

If the remaining columns are unique...

```julia
longdf = stack(iris, Not([:Species, :id]))
first(longdf, 3)
```

```julia
last(longdf, 3)
```

```julia
widedf = unstack(longdf, :variable, :value)
first(widedf, 3)
```

```julia
last(widedf, 3)
```

```julia
widedf = unstack(longdf) # even simpler
first(widedf, 3)
```

```julia
d = stack(iris, view=true)
first(d, 3)
```

```julia
d = stack(iris, Not(:Species))
first(d, 3)
```

```julia
x = by(d, [:variable, :Species], :value => mean => :vsum)
first(x, 3)
```

```julia
first(unstack(x, :Species, :vsum), 3)
```
