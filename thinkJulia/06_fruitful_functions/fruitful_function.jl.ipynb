```julia
function distance(x1, y1, x2, y2)
    dx = x2 - x1
    dy = y2 - y1
    dsq = dx^2 + dy^2
    @show dsq
    sqrt(dsq)
end
```

```julia
distance(0, 0, 3, 4)
```

```julia
function area(radius)
    π * radius^2
end
```

```julia
function circlearea(xc, yc, xp, yp)
    radius = distance(xc, yc, xp, yp) 
    area(radius)
end
```

```julia
# or
function circlearea(xc, yc, xp, yp)
    area(distance(xc, yc, xp, yp))
end
```
