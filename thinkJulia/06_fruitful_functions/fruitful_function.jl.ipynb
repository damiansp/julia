```julia
function distance(x1, y1, x2, y2)
    dx = x2 - x1
    dy = y2 - y1
    dsq = dx^2 + dy^2
    @show dsq
    sqrt(dsq)
end
```

```julia
distance(0, 0, 3, 4)
```

```julia
function fact(n)
    if n == 0
        return 1
    else
        recurse = fact(n - 1)
        result = n * recurse
        return result
    end
end
```

```julia
fact(7)
```

```julia
function fact(n)
    if n == 0
        return 1
    else
        return n * fact(n - 1)
    end
end
```

```julia
fact(8)
```

```julia
function fib(n)
    if n == 0
        return 0
    elseif n == 1
        return 1
    else
        return fib(n - 1) + fib(n - 2)
    end
end
```

```julia
fib(8)
```
