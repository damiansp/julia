```julia
function find(word, letter)
    index = firstindex(word)
    while index <= sizeof(word)
        if word[index] == letter
            return index
        end
        index = nextind(word, index)
    end
    -1
end
```

```julia
find("banana", 'n')
```

```julia
function countchars(str, char)
    counter = 0
    for letter in str
        if letter == char
            counter += 1
        end
    end
    counter
end
```

```julia
countchars("anna anna bo banna", 'a')
```
