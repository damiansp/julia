```julia
function printlyrics()
    println("I'm a lumberjack and I'm OK")
    println("I sleep all night and I work all day")
end
```

```julia
printlyrics()
```

```julia
function sing(times=1)
    for i = 1:times
        printlyrics()
    end
end
```

```julia
sing(4)
```
