### Variable-Length Arg Tuples

```julia
function printall(args...)
    println(args)
end
```

```julia
printall(1, 2.0, 3)
```

```julia
t = (7, 3)
```

```julia
divrem(t)
```

```julia
divrem(t...)
```

```julia
function sumall(args...)
    sum((args))
end
```

```julia
sumall(1, 2, 3)
```
