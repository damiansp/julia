```julia
t = [('a', 1), ('b', 2), ('c', 3)]
for (letter, number) in t
    println(number, ": ", letter)
end
```

```julia
function hasmatch(t1, t2)
    for (x, y) in zip(t1, t2)
        if x == y
            return true
        end
    end
    return false
end
```

```julia
function charfreq(text)
    chardict = Dict()
    for c in text
        if c in keys(chardict)
            chardict[c] += 1
        else
            chardict[c] = 1
        end
    end
    chardict
end
```
