```julia
function capitalizeall(t)
    res = []
    for s in t
        push!(res, uppercase(s))
    end
    res
end
```

```julia
capitalizeall(name)
```

```julia
function onlyupper(t)
    res = []
    for s in t
        if s == uppercase(s)
            push!(res, s)
        end
    end
    res
end
```

```julia
onlyupper("SamOwenSmith")
```

```julia
function nestedsum(t)
    sum = 0
    for i in t
        for j in i
            sum += j
        end
    end
    sum
end
```

```julia
t = [[1, 2], [3], [4, 5, 6]]
nestedsum(t)
```

```julia
function cumsum(arr)
    out = []
    total = 0
    for val in arr
        total += val
        push!(out, total)
    end
    out
end
```

```julia
cumsum([1, 2, 3, 4])
```

```julia
function arraytrim(arr)
    if length(arr) < 2
        return arr
    else
        return arr[2:(end - 1)]
    end
end
```

```julia
arraytrim([1, 2, 3, 4, 5])
```
