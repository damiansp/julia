```julia
function addall(t)
    total = 0
    for x in t
        total += x
    end
    total
end
```

```julia
a = 1:10
addall(a)
```

```julia
function capitalizeall(t)
    res = []
    for s in t
        push!(res, uppercase(s))
    end
    res
end
```

```julia
capitalizeall(name)
```

```julia
function onlyupper(t)
    res = []
    for s in t
        if s == uppercase(s)
            push!(res, s)
        end
    end
    res
end
```

```julia
onlyupper("SamOwenSmith")
```
