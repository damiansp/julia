```julia
struct Card
    suit::Int64
    rank::Int64
    
    function Card(suit::Int64, rank::Int64)
        @assert(1 ظëج suit ظëج 4, "suit must be between 1 and 4")
        @assert(1 ظëج rank ظëج 13, "rank must be between 1 and 13")
        new(suit, rank)
    end
end
```

```julia
function Base.show(io::IO, card::Card)
    print(io, RANKS[card.rank], SUITS[card.suit])
end
```

```julia
# Order by suits then face
function isless(c1::Card, c2::Card)
    (c1.suit, c1.rank) < (c2.suit, c2.rank)
end
```

### Decks

```julia
struct Deck1
    cards::Array{Card, 1}
end
```

```julia
function Deck1()
    deck = Deck1(Card[])
    for suit in 1:4
        for rank in 1:13
            push!(deck.cards, Card(suit, rank))
        end
    end
    deck
end
```

```julia
function Base.show(io::IO, deck::Deck1)
    for card in deck.cards
        print(io, card, " ")
    end
    println()
end
```

```julia
Deck1()
```

### Add, Remove, Shuffle & Sort

```julia
function Base.pop!(deck::Deck1)
    pop!(deck.cards)
end
```

```julia
function Base.push!(deck::Deck1, card::Card)
    push!(deck.cards, card)
    deck
end
```

```julia
function Random.shuffle!(deck::Deck1)
    shuffle!(deck.cards)
    deck
end
```

```julia
function Base.sort!(deck::Deck1)
    sort!(deck.cards)
    deck
end
```

```julia
deck = Deck1()
shuffle!(deck)
```

```julia
sort!(deck)
```

### Abstract Types and Subtyping

```julia
abstract type CardSet end
```

```julia
struct Deck <: CardSet
    cards::Array{Card, 1}
end
```

```julia
function Deck()
    deck = Deck(Card[])
    for suit in 1:4
        for rank in 1:13
            push!(deck.cards, Card(suit, rank))
        end
    end
    deck
end
```

```julia
deck = Deck()
deck isa CardSet
```

```julia
struct Hand <: CardSet
    cards::Array{Card, 1}
    label::String
end
```

```julia
function Hand(label::String="")
    Hand(Card[], label) # init as empty Array of Cards
end
```

```julia
hand = Hand("My hand")
```

```julia
function move!(cs1::CardSet, cs2::CardSet, n::Int)
    @assert 1 ظëج n ظëج length(cs1.cards)
    for i in 1:n
        card = pop!(cs1)
        push!(cs2, card)
    end
    nothing
end
```
