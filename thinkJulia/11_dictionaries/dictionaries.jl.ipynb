```julia
function histogram(s)
    d = Dict()
    for c in s
        if !(c in keys(d))
            d[c] = 1
        else
            d[c] += 1
        end
    end
    d
end
```

```julia
h = histogram("antidiestablishmentarianism")
```

```julia
get(h, 'i', 0)
```

```julia
get(h, 'z', 0)
```

```julia
function printhist(h)
    for c in keys(h)
        print(c, " ")
        for i in 1:h[c]
            print("*")
        end
        println()
    end
end
```

```julia
printhist(h)
```

```julia
function reverselookup(d, v)
    for k in keys(d)
        if d[k] == v
            return k
        end
    end
    error("LookupError")
end
```

```julia
h = histogram("parrot")
```

```julia
reverselookup(h, 2)
```

```julia
reverselookup(h, 1)
```

```julia
function invertdict(d)
    inverse = Dict()
    for k in keys(d)
        v = d[k]
        if !(v in keys(inverse))
            inverse[v] = [k]
        else
            push!(inverse[v], k)
        end
    end
    inverse
end
        
```

```julia
inverse = invertdict(h)
```

# Memos

```julia
known = Dict(0 => 0, 1 => 1)

function fibonacci(n)
    # memoized version
    if n in keys(known)
        return known[n]
    end
    res = fibonacci(n - 1) + fibonacci(n - 2)
    known[n] = res
    res
end
```
