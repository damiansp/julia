```julia
function countdown(n)
    if n ≤ 0
        println("Blastoff!")
    else
        print(n, "... ")
        countdown(n - 1)
    end
end
```

```julia
countdown(7)
```

```julia
function print_n_times(s, n)
    if n <= 0
        return
    end
    println(s)
    print_n_times(s, n - 1)
end
```

```julia
print_n_times("Hey, hey. Hello, Echo!", 3)
```

```julia
function time_since_epoch()
    seconds = time()
    minutes = seconds ÷ 60
    seconds %= 60
    hours = minutes ÷ 60
    minutes %= 60
    days = hours ÷ 24
    hours %= 24
    years = days ÷ 365.25
    days %= 365.25
    ("$(Int32(years)) years, $(Int32(round(days))) days, $(Int32(hours)) "
     * "hours, $(Int32(minutes)) minutes, $(Int32(round(seconds))) "
     * "seconds")
end
```

```julia
time_since_epoch()
```

```julia
function istriangle(a, b, c)
    if (a > b + c) || (b > a + c) || (c > a + b)
        return false
    end
    return true
end
```

```julia
istriangle(1, 1, 1)
```

```julia
istriangle(3, 1, 1)
```
