```julia
function countdown(n)
    if n ≤ 0
        println("Blastoff!")
    else
        print(n, "... ")
        countdown(n - 1)
    end
end
```

```julia
countdown(7)
```

```julia
function print_n_times(s, n)
    if n <= 0
        return
    end
    println(s)
    print_n_times(s, n - 1)
end
```

```julia
print_n_times("Hey, hey. Hello, Echo!", 3)
```
