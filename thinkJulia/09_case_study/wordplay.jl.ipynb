```julia
function findiflength(file, minlen)
    for line in eachline(file)
        line = replace(line, " " => "")
        if length(line) >= minlen
            println(line)
        end
    end
end            
```

```julia
findiflength("../../data/words.txt", 20)
```

```julia
function hasno_e(str)
    for c in str
        if c == 'e' || c == 'E'
            return false
        end
    end
    return true
end
```

```julia
hasno_e("Damian"), hasno_e("Fred"), hasno_e("Edwina")
```

```julia
function percent_no_e(file)
    n = 0
    no_es = 0
    for line in eachline(file)
        n += 1
        if hasno_e(line)
            no_es += 1
        end
    end
    100 * no_es / n
end
```

```julia
percent_no_e("../../data/words.txt")
```

```julia
function usesonly(str, allowable)
    for c in str
        if !(c in allowable)
            return false
        end
    end
    true
end
```

```julia
usesonly("alfalfa", "alf")
```

```julia
usesonly("flap", "alf")
```

```julia
function words_usingonly(file, allowable)
    for line in eachline(file)
        if usesonly(line, allowable)
            print(line, " ")
        end
    end
end
```

```julia
words_usingonly("../../data/words.txt", "acefhlo")
```

```julia
function usesall(str, required)
    for c in required
        if !(c in str)
            return false
        end
    end
    true
end
```

```julia
function uses_allvowels(file; include_y=false)
    required = include_y ? "aeiouy" : "aeiou"
    for line in eachline(file)
        if usesall(line, required)
            print(line, " ")
        end
    end
end
```

```julia
uses_allvowels("../../data/words.txt"; include_y=true)
```
