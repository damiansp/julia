```julia
function walk(dirname)
    for name in readdir(dirname)
        path = joinpath(dirname, name)
        if ispath(path)
            println(path)
        else
            walk(path)
        end
    end
end
```

```julia
fin = open("/etc/passwd", "w")
```

```julia
try
    fin = open("nonesuch")
catch exc
    println("No can do: $exc")
end
```

```julia
f = open("output.txt")
try
    line = readline(f)
    println(line)
catch exc
    println("$exc")
finally
    close(f)
end
```
