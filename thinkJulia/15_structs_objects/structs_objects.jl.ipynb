```julia
x = p.x
y = p.y
x, y
```

```julia
distance_from_origin = sqrt(p.x^2 + p.y^2)
distance_from_origin
```

### Mutable structs

```julia
mutable struct MPoint
    x
    y
end
```

```julia
shifty = MPoint(0., 0.)
```

```julia
shifty.x = 3.
shifty.y = 4.
```

### Rectangles

```julia
struct Rectangle
    width
    height
    corner
end
```

```julia
origin = MPoint(0., 0.)
box = Rectangle(100., 200., origin)
```

```julia
function printpoint(p)
    println("($(p.x), $(p.y))")
end
```

```julia
printpoint(shifty)
```

```julia
function getdist(p1, p2)
    sqrt((p1.x - p2.x)^2 + (p1.y - p2.y)^2)
end
```

```julia
getdist(shifty, origin)
```

```julia
function movepoint!(p, dx, dy)
    p.x += dx
    p.y += dy
    nothing
end
```

```julia
movepoint!(shifty, -4, 7)
printpoint(shifty)
```

```julia
function moverect!(rect, dx, dy)
    movepoint!(rect.corner, dx, dy)
end
```

```julia
box
```

```julia
moverect!(box, 1., 2.)
box
```

```julia
box.corner = MPoint(1., 2.)
```

### Instances as Return Values

```julia
function findcenter(rect)
    Point(rect.corner.x + rect.width / 2, rect.corner.y + rect.height / 2)
end
```

```julia
center = findcenter(box)
printpoint(center)
```

```julia
struct Circle
    r
    c
end
```

```julia
c1 = Circle(75, Point(150, 100))
```

```julia
function pointincircle(point, circle)
    dist = getdist(point, circle.c)
    dist <= circle.r
end
```
